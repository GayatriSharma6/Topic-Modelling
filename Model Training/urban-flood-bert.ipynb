```python
from flair.embeddings import TransformerDocumentEmbeddings
from bertopic import BERTopic

base_cased = TransformerDocumentEmbeddings('../input/bert-base-uncased')
```

```python
docs=df['Text'].values
new_docs=[]
for i in range(len(docs)):
    new_docs.append(str(docs[i]))
```

```python

topic_model = BERTopic(embedding_model=base_cased, language="english",
                       calculate_probabilities=False, verbose=True,nr_topics='auto'
                      )

topics,_=topic_model.fit_transform(new_docs)
```
